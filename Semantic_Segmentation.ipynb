```
# Arguments
BBOX = '-73.983958,40.740868,-73.975911,40.745664'
OUTPUT_DIR = 'data/output_images'
SIZE = 1024  # {320, 640, 1024, 2048}
```

```
# Modules
import requests
from pathlib import Path
```

```
Path(OUTPUT_DIR).mkdir(parents=True, exist_ok=True)
```

```
# Making a Request
url = f'https://a.mapillary.com/v3/images?bbox={BBOX}&client_id={ID}&per_page=1000'
headers = {'Authorization': f'Bearer {TOKEN}'}
response = requests.get(url, headers=headers)
```

```
# print(response.json())
data = response.json()['features'][0]
print(data)

coords = data['geometry']['coordinates']
img_key = data['properties']['key']
print(coords)
# print(img_key)
```

```
# THIS CODE WILL GET MANY IMAGES AT THE SAME TIME. IF YOU GET 'INDEX OUT OF RANGE ERROR' - DECREASE THE NUMBER_OF_IMAGES VALUE. THERE ARE LESS PHOTOS, PROBABLY, IN THE AREA YOU'RE LOOKING AT.

new_data = []
new_coords = []
new_ca = []
new_image_key = []
NUMBER_OF_IMAGES = 1000

for i in range (0, NUMBER_OF_IMAGES):
  new_data.append(response.json()['features'][i])
  new_coords.append(new_data[i]['geometry']['coordinates'])
  new_ca.append(new_data[i]['properties']['ca'])
  new_image_key.append(new_data[i]['properties']['key'])

# create new dir for data for gh
GH_DATA_DIR = 'data/gh_data'
Path(GH_DATA_DIR).mkdir(parents=True, exist_ok=True)

# write ca (camera angle) to ca.txt file at gh dir
with open(f'{GH_DATA_DIR}/ca.txt', 'w') as f:
  for j in range (0, len(new_ca)-1):
    f.write(str(new_ca[j]) + "\n")

# write camera lng coords to lng_coords.txt file at gh dir
with open(f'{GH_DATA_DIR}/lng_coords.txt', 'w') as f1:
  for j in range (0, len(new_coords)-1):
    f1.write(str(new_coords[j][0]) + "\n")

# write camera ltd coords to ltd_coords.txt file at gh dir
with open(f'{GH_DATA_DIR}/ltd_coords.txt', 'w') as f2:
  for j in range (0, len(new_coords)-1):
    f2.write(str(new_coords[j][1]) + "\n")

# retrieving images
new_image_url = []
new_image_req = []
for k in range(0, NUMBER_OF_IMAGES):
  new_image_url.append(f'https://images.mapillary.com/{new_image_key[k]}/thumb-{SIZE}.jpg')
  new_image_req.append(requests.get(new_image_url[k]))

for l in range (0, len(new_image_req)-1):
  FILE_NAME = f'{OUTPUT_DIR}/output_img' + str(l) + f'.jpg'
  with open(FILE_NAME, 'wb') as f3:
    f3.write(new_image_req[l].content)
```
